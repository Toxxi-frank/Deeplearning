```python
from tensorflow.keras.applications.inception_v3 import InceptionV3
# import mô hình InceptionV3 do dùng Transfer Learning
```

```python
local_weights_file = './inception_v3_weights_tf_dim_ordering_tf_kernel_nostop.h5'
```

```python
import os
from tensorflow.keras import layers
from tensorflow.keras import Model

from tensorflow.keras.applications.inception_v3 import InceptionV3
```

```python
pre_trained_model = InceptionV3(input_shape = (150,150,3),
                                include_top=False, weights = None)
# include_top dùng để bỏ đi lớp FC (softmax) trên cùng, thấy trùng với pre_trained model no_top
```

```python
pre_trained_model.load_weights(local_weights_file)
# load model weigth đã được trained
```

```python
pre_trained_model.summary()
# xem cấu trúc trained_model
```

```python
for layer in pre_trained_model.layers:
  layer.trainable = False
# đóng băng cac layers dùng để train
```

```python
pre_trained_model.summary()
```

Thấy ở trên Trainable params = 0

```python
last_layer = pre_trained_model.get_layer('mixed7')
# lấy ra layer tên là mixed7 
```

```python
last_output = last_layer.output
```

```python
last_output
```

```python
from tensorflow.keras.optimizers import RMSprop
# import cái optimizer lựa chọn để train model 
```

```python
x = layers.Flatten()(last_output)
#Flatten the output layer to 1 dimension
```

```python
x = layers.Dense(1024, activation = 'relu')(x)
# add fc layer with 1024 hidden units and relu activation
```

```python
x = layers.Dropout(0.2)(x)
# add dropout rate 0.2
```

```python
x = layers.Dense(1, activation = 'sigmoid')(x)
# add a final activation for classification
```

```python
model = Model(pre_trained_model.input, x)
```

```python
model.compile(optimizer = RMSprop(learning_rate= 0.0001), loss = 'binary_crossentropy', metrics = ['acc'])
```

```python
model.summary()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
# xử lí hình ảnh, tăng thêm data 
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255., rotation_range = 40,
                                   width_shift_range =0.2, height_shift_range =0.2, shear_range=0.2,
                                   zoom_range = 0.2, horizontal_flip=True)
```

```python
vadilation_datagen = ImageDataGenerator(rescale = 1./255.)
```

```python
train_generator = train_datagen.flow_from_directory(train_folder, batch_size =20, class_mode='binary', target_size=(150,150))
```

```python
validation_generator = vadilation_datagen.flow_from_directory(valid_folder, batch_size=20, class_mode='binary', target_size=(150,150))
```

```python
train_generator.class_indices
# dùng để xác định class_mode cho các trên (thấy 2 labels --> binary)
```

```python
history = model.fit_generator(train_generator, validation_data= validation_generator, steps_per_epoch=50,epochs=20, validation_steps=50, verbose=2 )
```
